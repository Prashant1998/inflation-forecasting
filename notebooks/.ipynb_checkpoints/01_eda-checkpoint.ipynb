```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (10,5)
```

```python
cpi = pd.read_csv("../data/raw/cpi.csv")
cpi.columns = ["date", "cpi"]
cpi["date"] = pd.to_datetime(cpi["date"])

```

```python
rate = pd.read_csv("../data/raw/interest_rate.csv")
rate.columns = ["date", "interest_rate"]
rate["date"] = pd.to_datetime(rate["date"])
```

```python
unemp = pd.read_csv("../data/raw/unemployment.csv")
unemp.columns = ["date", "unemployment"]
unemp["date"] = pd.to_datetime(unemp["date"])
```

```python
ip = pd.read_csv("../data/raw/industrial_production.csv")
ip.columns = ["date", "industrial_production"]
ip["date"] = pd.to_datetime(ip["date"])
```

```python
df = cpi.merge(rate, on="date", how="inner")
df = df.merge(unemp, on="date", how="inner")
df = df.merge(ip, on="date", how="inner")
```

```python
df = df.set_index("date")
df.head()
```

```python
df.index.inferred_freq
df.isna().sum()
```

```python
df = df.dropna()

```

```python
df.to_csv("../data/processed/us_macro_monthly.csv")
```

Data Construction Note
Macroeconomic indicators are released at different times.
An inner join is used to ensure temporal consistency and avoid leakage
from future information.
