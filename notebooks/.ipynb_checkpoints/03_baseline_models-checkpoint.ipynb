```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.metrics import mean_absolute_error
```

```python
df = pd.read_csv(
    "../data/processed/macro_stationary.csv",
    parse_dates=["date"]
)
df = df.set_index("date")

df.head()
```

```python
y = df["inflation_rate"]
```

```python
split_date = "2015-01-01"

train = y[y.index < split_date]
test  = y[y.index >= split_date]
```

BASELINE 1: NAIVE(LAG=1)



```python
naive_forecast = test.shift(1)
```

```python
naive_forecast = naive_forecast.dropna()
test_naive = test.loc[naive_forecast.index]
```

```python
print(
    "Naive MAE:",
    mean_absolute_error(test_naive, naive_forecast)
)
```

```python
plt.plot(test_naive, label="Actual")
plt.plot(naive_forecast, label="Naive")
plt.legend()
plt.title("Naive Baseline Forecast")
plt.show()
```

 BASELINE 2: ROLLING MEAN

```python
rolling_mean = train.rolling(window=12).mean().iloc[-1]
rolling_forecast = pd.Series(
    rolling_mean,
    index=test.index
)
```

```python
print(
    "Rolling Mean MAE:",
    mean_absolute_error(test, rolling_forecast)
)
```

BASELINE 3 : ARIMA

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
model = ARIMA(train, order=(1,0,1))
model_fit = model.fit()
```

```python
arima_forecast = model_fit.forecast(steps=len(test))
arima_forecast.index = test.index
```

```python
print(
    "ARIMA MAE:",
    mean_absolute_error(test, arima_forecast)
)
```

Baseline Results

Naive and ARIMA models set a strong performance floor.
This highlights the persistence and limited signal content of inflation data.
