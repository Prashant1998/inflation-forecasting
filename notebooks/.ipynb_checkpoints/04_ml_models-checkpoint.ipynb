```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.metrics import mean_absolute_error
```

```python
df = pd.read_csv(
    "../data/processed/macro_stationary.csv",
    parse_dates=["date"]
)
df = df.set_index("date")

df.head()
```

```python
y = df["inflation_rate"]
X = df.drop(columns=["inflation_rate"])
```

```python
for lag in [1, 3, 6, 12]:
    X[f"inflation_lag_{lag}"] = y.shift(lag)
```

```python
data = pd.concat([y, X], axis=1).dropna()

y = data["inflation_rate"]
X = data.drop(columns=["inflation_rate"])
```

```python
split_date = "2015-01-01"

X_train = X[X.index < split_date]
X_test  = X[X.index >= split_date]

y_train = y[y.index < split_date]
y_test  = y[y.index >= split_date]
```

```python
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled  = scaler.transform(X_test)
```

MODEL 1 = LINEAR REFGRESSION

```python
lr = LinearRegression()
lr.fit(X_train_scaled, y_train)

lr_pred = lr.predict(X_test_scaled)

print(
    "Linear Regression MAE:",
    mean_absolute_error(y_test, lr_pred)
)
```

MODEL 2: RIDGE REGRESSION

```python
ridge = Ridge(alpha=1.0)
ridge.fit(X_train_scaled, y_train)

ridge_pred = ridge.predict(X_test_scaled)

print(
    "Ridge Regression MAE:",
    mean_absolute_error(y_test, ridge_pred)
)
```

ML vs Baselines

Linear models with lag features show limited improvement over naive and ARIMA baselines, indicating that most predictive power comes from persistence rather than rich exogenous structure.

```python
plt.plot(y_test.index, y_test, label="Actual")
plt.plot(y_test.index, lr_pred, label="Linear Regression")
plt.legend()
plt.title("ML Forecast vs Actual")
plt.show()
```

Limitations

Small sample size

Structural breaks and policy regime shifts

Linear ML models cannot capture nonlinear macro dynamics
